```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
titanic_df=pd.read_csv("titanic.csv")
```

```python
titanic_df.head()
```

```python
le=LabelEncoder()
titanic_df.Sex=le.fit_transform(titanic_df.Sex)

titanic_df.Sex.head()
```

```python
titanic_df.describe()
```

```python
titanic_df.head()
```

```python
pclass_dummies = pd.get_dummies(titanic_df["Pclass"])
pclass_dummies.columns=["Pclass_1","Pclass_2","Pclass_3"]
sex_dummies=pd.get_dummies(titanic_df["Sex"])
sex_dummies.columns=["Male","Female"]
PC_dummies=pd.get_dummies(titanic_df["Parents/Children Aboard"])
PC_dummies.columns=["PC1","PC2","PC3","PC4","PC5","PC6","PC7"]
SSA_dummies=pd.get_dummies(titanic_df["Siblings/Spouses Aboard"])
SSA_dummies.columns=["SSA1","SSA2","SSA3","SSA4","SSA5","SSA6","SSA7"]
```

```python
titanic_df = pd.concat([titanic_df,pclass_dummies,sex_dummies,PC_dummies,SSA_dummies],axis=1)
```

```python
titanic_df.head()
```

```python
titanic_df=titanic_df.drop(["Name","Pclass","Siblings/Spouses Aboard","Parents/Children Aboard","Sex"],axis=1)
titanic_df
```

```python
# decision tree implementation after data cleaning
```

```python
# seperating independent and dependent variable
x=titanic_df.drop("Survived",axis=1)
y=titanic_df["Survived"]
```

```python
# import Model selector
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=101,stratify=y)
```

```python
train_y.value_counts()/len(train_y)
```

```python
test_y.value_counts()/len(test_y)
```

```python
# since the output values are same we can say that the test and train data are similar
```

```python
#Decision classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
clf.fit(train_x,train_y)
```

```python
clf.score(train_x,train_y)
```

```python
clf.score(test_x,test_y)
```
